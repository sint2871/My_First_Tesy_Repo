```python
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

files=files.upload() #Загружаем файл в библиотеку Python
Z = np.loadtxt("f.txt", delimiter='\t', dtype=np.float64) #Функция np для вывода из файла массива np.array
#Суммарные затраты по 1 тарифу
s1=[3.08]*24 #запись в виде массива 1 тариф
Z1=np.multiply(s1,Z) #Перемножение массивов
S1=sum(Z1) #Сумма массива
print ('Расходы по 1 тарифу = ', S1, 'руб')

#Суммарные затраты по 2 тарифу
s2=[2.40]*7+[3.54]*16+[2.40]*1 #запись в виде массива 2 тариф
Z2=np.multiply(s2,Z)#Перемножение массивов
S2=sum(Z2)#Сумма массива
print ('Расходы по 2 тарифу = ', S2, 'руб')
#Суммарные затраты по 3 тарифу
s3=[2.40]*7+[3.80]*3+[3.08]*7+[3.80]*4+[3.08]*2+[2.40]*1 #запись в виде массива 3 тариф
Z3=np.multiply(s3,Z)#Перемножение массивов
S3=sum(Z3)#Сумма массива
print ('Расходы по 3 тарифу = ', S3, 'руб')
#Поиск минимального значения по тарифу
m=min(S1,S2,S3)
if m==S1:
    print("Самый выгодный первый тариф")
else:
    if m==S2:
      print("Самый выгодный второй тариф")
    else:
      if m==S3:
        print("Самый выгодный третий тариф")
```
